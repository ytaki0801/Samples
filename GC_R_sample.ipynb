```
x <- seq(-3, 3, length=50)
y <- x
r <- 0.9
f <- function(x, y) {
  1/(2*pi*sqrt(1-r^2))*exp(-(x^2-2*r*x*y+y^2)/(2*(1-r^2)))
}
z <- outer(x, y, f)
z[is.na(z)] <- 1
persp(x, y, z)
```
